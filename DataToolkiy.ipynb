1. What is NumPy, and why is it widely used in Python?

NumPy (Numerical Python) is a Python library used for high-performance numerical computing.
It provides:

n-dimensional array objects (ndarray)

Fast mathematical operations

Vectorization

Linear algebra, Fourier transforms, statistics

Memory-efficient storage

Why it is widely used:

Written in C, so operations are super fast.

Supports array broadcasting.

Foundation library for Pandas, Scikit-learn, TensorFlow, PyTorch.

Handles large datasets efficiently.

2. How does broadcasting work in NumPy?

Broadcasting allows NumPy to operate on arrays of different shapes by automatically expanding the smaller array.

Example:

[1, 2, 3] + 5 → [6, 7, 8]


Rules:

Compare shapes from right to left.

Dimensions must be equal, or one of them is 1, or missing.

NumPy stretches the array with dimension 1.

Broadcasting avoids loops → faster and cleaner code.

4. Explain the use of the groupby() method in Pandas.

groupby() is used to split data into groups based on a column, apply a function, and combine results.

Example:

df.groupby("gender")["salary"].mean()


It performs:

Split → by column

Apply → function (sum, mean, count)

Combine → returns summarized output

Useful for:

Aggregation

Summarization

Analytics

21. Why is NumPy’s array slicing faster than Python list slicing?

Because:

NumPy uses views (no data copy).

Stored in contiguous memory, enabling C-level pointer arithmetic.

Python lists store objects → more overhead.
